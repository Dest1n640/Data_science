```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
y = np.random.rand(100000)
plt.figure(figsize=(3, 3))
plt.hist(y, bins = 300)
plt.show()
```

```python
x1 = np.random.rand(10000)
x2 = np.random.rand(10000)
plt.figure(figsize = (5, 5))
plt.hist(x1 + x2, bins = 300)
plt.show()
```

```python
x1 = np.random.rand(100000)
x2 = np.random.rand(100000)
x3 = np.random.rand(100000)
plt.figure(figsize=(4, 4))
plt.hist(x1 + x2 + x3, bins = 300)
plt.show()
```

```python
x = np.random.rand(100000, 20)
y = x.sum(axis=-1)
plt.hist(y, bins = 300)
plt.show()
y.mean() ,y.std(), y.var()
```

```python
x.mean() * 20, x.std() * np.sqrt(20)
```

```python
y = x.mean(axis = -1)
x.mean(), x.std()/np.sqrt(20), y.mean(), y.std()
```

```python
import scipy.stats as ss
y = x.mean(axis= -1)
t = ss.norm(x.mean(), x.std()/np.sqrt(20)).rvs(100000)
plt.hist(y, bins = 300)
plt.hist(t, bins=300, histtype='step')
```

```python
import scipy.stats as ss
x = ss.gibrat(10).rvs((100000, 20))
y = x.mean(axis= -1)
t = ss.norm(x.mean(), x.std()/np.sqrt(20)).rvs(100000)
plt.hist(y, bins = 300)
plt.hist(t, bins=300, histtype='step')
```

```python
import scipy.stats as ss
x = ss.norm(0, 1).rvs((100000, 2))
t = ss.chi2(df = 2).rvs((100000))
y = (x**2).sum(axis= -1)
plt.hist(y, bins = 300)
plt.hist(t, bins = 300, histtype='step')
```
