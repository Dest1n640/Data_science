```python
import scipy.stats as ss 
dist = ss.bernoulli(0.5)
x =dist.rvs(20)
x
plt.hist(x)
```

```python
import scipy.stats as ss 
dist = ss.bernoulli(0.25)
x = dist.rvs((10000,10))
x.shape
y = x.sum(axis = -1)
plt.figure(figsize=(5,5))
plt.hist(y, bins=100)
```

```python
dist = ss.binom(10, 0.25)
y = dist.rvs(1000)
plt.Figure(figsize=(3, 3))
plt.hist(y, bins = 100);
```

```python
dist = ss.poisson(3)
y = dist.rvs(1000)
plt.Figure(figsize=(3, 3))
plt.hist(y, bins = 100);
```
