```python
import numpy as np
import matplotlib.pyplot as plt
f = open('fileXYZ.dat')
d = []
for s in f:
  d.append([float(x) for x in s.split()])
d = np.array(d)
fig, axs = plt.subplots(1, 3, figsize = (10, 3), constrained_layout = True)
axs[0].plot(d[:,0], label = 'x(t)')
axs[0].legend()
axs[1].plot(d[:,1], label = 'y(t)')
axs[1].legend()
#axs[2].plot(d[:,0], d[:, 2], label = 'z(x)')
axs[2].legend()
axs[2].scatter(d[:,0],d[:,2],label = 'z(x)') #    
axs[0].set_xlim(0, 20)
axs[1].set_xlim(200,300)
axs[2].set_xlabel('x')
axs[2].set_ylabel('z')
axs[0].set_xlabel('t')
axs[0].set_ylabel('x')
axs[1].set_xlabel('t')
axs[1].set_ylabel('y')
```
