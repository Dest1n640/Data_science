```python
import numpy as np
import matplotlib.pyplot as plt 
import scipy.stats as ss 
```

```python
x = ss.gompertz(10).rvs(100000, 20)
y = x.mean(axis=-1)
plt.hist(x, bins = 300)
#plt.hist(y, bins=300)
```

```python
d = ss.gompertz(10)
plt.hist(x, bins = 300, density=True, cumulative=True)
t = np.linspace(x.min(), x.max(), 1000)
plt.plot(t, d.cdf(t))
```

```python
d = ss.gompertz(10)
plt.hist(x, bins = 300, density=True)
t = np.linspace(x.min(), x.max(), 1000)
plt.plot(t, d.pdf(t))
```

```python
xL, xH = d.ppf(0.025), d.ppf(0.975)
print(xL, xH)
y = x[x >=xL]
y = y[y <= xH]
print(y.shape[0] / x.shape[0]) #
```

```python
xL, xH = np.percentile(x, 2.5), np.percentile(x, 97.5)
y = x[x >= xL]
y = y[y <= xH]
print(y.shape[0] / x.shape[0]) #
```

```python
normal = ss.norm(0, 1)
normal.ppf(0.025), normal.ppf(0.975)
```
